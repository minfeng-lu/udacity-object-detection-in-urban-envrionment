## Write a function to display an image and the bounding boxes

Implement the `display_instances` function below. This function takes a batch as an input and display an image with its corresponding bounding boxes. The only requirement is that the classes should be color coded (eg, vehicles in red, pedestrians in blue, cyclist in green).

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from PIL import Image
%matplotlib inline

def display_instances(batch):
    """
    This function takes a batch from the dataset and display the image with 
    the associated bounding boxes.
    """
    # ADD CODE HERE
    plt.figure(figsize=(20,20))
    color_map = {1: [1, 0, 0], 2: [0, 1, 0], 3: [0, 0, 1]}
    for idx, data in enumerate(batch):
        row = int(idx / 4)
        col = idx % 4
        image = Image.fromarray(data['image'].numpy())
        ax = plt.subplot(5, 4, idx + 1)
        original_image_spatial_shape = data["original_image_spatial_shape"].numpy()
        w = original_image_spatial_shape[0]
        h = original_image_spatial_shape[1]
        for box, label in zip(data['groundtruth_boxes'], data['groundtruth_classes']):
            y1, x1, y2, x2 = box.numpy() * [w, h, w, h]
            l = label.numpy()
            rec = Rectangle((x1, y1), x2-x1, y2-y1, facecolor = "none", edgecolor=color_map[l], linewidth=2)
            ax.add_patch(rec)
        ax.imshow(image, aspect='auto')
        ax.axis("off")
    plt.tight_layout()     
    plt.show()
```

## Display 10 images 

Using the dataset created in the second cell and the function you just coded, display 10 random images with the associated bounding boxes. You can use the methods `take` and `shuffle` on the dataset.

```python
## STUDENT SOLUTION HERE
for i in range(10):
    # randonly shuffle from a buffer of size 100
    batch = dataset.shuffle(100, reshuffle_each_iteration=True).take(1)
    display_instances(batch)
```

## Additional EDA

In this last part, you are free to perform any additional analysis of the dataset. What else would like to know about the data?
For example, think about data distribution. So far, you have only looked at a single file...

```python
label_map = {}
batches = dataset.shuffle(100, reshuffle_each_iteration=True).take(100)
for batch in batches:
    groundtruth_classes = batch["groundtruth_classes"].numpy()
    for label in groundtruth_classes:
        count = label_map.get(label, 0)
        count += 1
        label_map[label] = count
            
print(label_map)        
data = {"car": label_map.get(1, 0), "pedestrian": label_map.get(2, 0), "cyclist": label_map.get(4, 0)}
keys = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
plt.bar(keys, values, width = 0.4)
 
plt.xlabel("Label")
plt.ylabel("Count")
plt.title("Class Distribution")
plt.show()
```
